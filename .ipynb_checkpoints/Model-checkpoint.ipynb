## Remove Citation sources at the beg of every real text

### Real text: finding real text with no citation
By this dataset design, citation of source (Reuters) always comes before a hyphen

If there is no hyphen => no citation

Twitter tweets always have the same sources, along text of max 280 char, follow by a hyphen

```python
real_data.sample(5) 
```

```python
#create a for loop to find texts with no citation of publications
unknown_publishers_index = [] 
for index, elements in enumerate(real_data.text.values): 
    try: 
        record = elements.split(" - ", maxsplit = 1)
        record[1]      
        assert(len(record[0])<280) 

    except:
        unknown_publishers_index.append(index) 
```

```python
# view the unknown_publisher
unknown_publishers_index
```

```python
len(unknown_publishers_index)
```

```python
#view the texts with no source citation
real_data.iloc[unknown_publishers_index].head(15)
```

### Real text: remove publisher information from text (now that we know which text do not have citation of source) & create clean text (no citation within)

```python
publisher =[] # list to store the source name
tmp_text = [] # to store the text (no citation)

for index, elements in enumerate(real_data.text.values):
    if index in unknown_publishers_index:
        tmp_text.append(elements)
        publisher.append("Unknown") 
        
    else: 
        record = elements.split(" - ", maxsplit=1) 
        publisher.append(record[0]) 
        tmp_text.append(record[1])
        
```

```python
real_data2 = real_data.copy()
real_data2.head()
```

```python
real_data2["text"] = tmp_text #put the tmp_text list into the [text] column
real_data2["publisher"] = publisher #.Create publisher column. put the pub list into the publisher column
real_data2.head()
```

```python
real_data2.shape
```

### Merge title & text & lowercase all the words

```python
#merge title & text
real_data2 ["text"] = real_data2 ["title"] + " " + real_data2 ["text"]   
fake_data ["text"] = fake_data ["title"] + " " + fake_data ["text"]
```

```python
real_data2["text"] = real_data2["text"].apply(lambda x: str(x).lower())
fake_data["text"] = fake_data["text"].apply(lambda x: str(x).lower())
```

# 3. Pre-process Data

### Create a X --> Y Mapping dataset.

```python
# Create class column for 2 dataframe
real_data2["class"] = 1
fake_data["class"] = 0
```

```python
real_data2.columns
#we only need the text & the class columns for model training
```

```python
#create x-y dataset (also a dataframe)
real_set = real_data2[["text", "class"]]
fake_set = fake_data[["text", "class"]]
```

```python
real_set.head()
```

```python
fake_set.head()
```

```python
#append the 2 dataframes together
dataset = real_set.append(fake_set, ignore_index=True)
dataset.sample(16)
```

### Remove special characters

```python
#using regex
dataset["text"] = dataset["text"].str.replace('[^A-Za-z0-9\s]', '') 
dataset["text"] = dataset["text"].str.replace('\n', '')
dataset["text"] = dataset["text"].str.replace('\s+', ' ')
```

```python
dataset.head()
```

### Remove Stop Words

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```python
dataset['text'] = dataset['text'].apply(lambda x: " ".join([word for word in x.split() if word not in stop]))
dataset.head()
```

### Vectorization: convert text data to numerical data
Using Word2Vec (word to vector): a popular technique to learn word embeddings using shalow NN, from that we have a word vector.

Word embedding: a term used for the representation of words, in the form of a real-valued vector that encodes the meaning of the. Words that are closer in the vector space are expected to be similar in meaning. In a sense, each vector value represent a particular meaning.

These representation is capabe of capturing context of a word in a document, in relation to other words.

```python
#use gensim module to create word-embeddings
import gensim
```

```python
y = dataset["class"].values
y 
```

```python
X = [d.split() for d in dataset["text"].tolist()]
```

```python
type(X), type(y)
```

```python
#convert words into word-embeddings

DIM = 100 

#create a Word2Vec model object & train 
w2v_model = gensim.models.Word2Vec(sentences=X, vector_size = DIM, window=10, min_count=1)
```

```python
#https://stackoverflow.com/questions/35596031/gensim-word2vec-find-number-of-words-in-vocabulary
vocab_len = len(w2v_model.wv)
vocab_len
```

```python
#view the vector attribute of some words (keys)
w2v_model.wv["love"]
#100 vectors within
```

```python
w2v_model.wv["china"]
```

```python
#find the most similar word to "china", based on vector property
w2v_model.wv.most_similar("china")
#like, if you mention china, the countries below are also mentioned quite often.
```

```python
w2v_model.wv.most_similar("us")
```

### Tokenize our words

```python
tokenizer = Tokenizer() 
tokenizer.fit_on_texts(X) 
```

```python
#tokenize our text
X = tokenizer.texts_to_sequences(X)
```

```python
#how many sequences in the list X
len(X)
```

```python
#see length of each sequence
number_of_sequence = np.array([len(x) for x in X])
number_of_sequence
```

```python
#see WHICH sequence have >1000 words
number_of_sequence[number_of_sequence>900]
```

```python
#see how many sequence have >1000 words
len(number_of_sequence[number_of_sequence>1000])
```

### Padding: 
To make all sequence have the same length

```python
maxlen = 900
X = pad_sequences(X,maxlen=maxlen)
```

```python
#check the length of each sequence, all are 900
np.array([len(x) for x in X])
```

### Create vocab size (with unknown words consideration)
When we input a news for the model to classify, there will be words that is not in this dataset

==> the model will haven't seen those words ==> need a way to tokenize them into sequence
==> we will consider adding an extra token as unknown words 
==> our current token numbers + 1 extra token for unknown words = our vocab_size

```python
#create vocab size
vocab_size = len(tokenizer.word_index) + 1
current_vocab = tokenizer.word_index #a dictionary
```

### Weight vectors

```python
#define a method to transform these word vectors into matrix

def get_weight_matrix(model):
    weight_matrix = np.zeros((vocab_size, DIM)) #return 2D array, shape (192216, 100)
    for word, index in current_vocab.items():
        weight_matrix[index] = model.wv[word]
    
    return weight_matrix    
```

```python
embedding_vectors = get_weight_matrix(w2v_model)
embedding_vectors.shape, type(embedding_vectors) #2D array
```

# 4. Create & train model

### Build the model

```python
model = Sequential()
model.add(Embedding(vocab_size, output_dim=DIM, weights=[embedding_vectors], input_length=maxlen, trainable=False))
model.add(LSTM(units=128))
model.add(Dense(1, activation="sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["acc"])
```

```python
model.summary()
```

### Train, test split

```python
# from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

### Train model

```python
#train model & create a History object for later plot
history = model.fit(X_train, y_train, validation_split = 0.15, batch_size=256, epochs = 8 )
```

# 5. Evaluate model

```python
# visualize the training process
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend(['Train', 'Validation'])
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend(['Train', 'Validation'])
plt.show()
```

```python
#make predictions on test set
y_pred = (model.predict(X_test) >= 0.5).astype(int)
```

```python
accuracy_score(y_test, y_pred)
#98.76% accuracy
```

```python
print(classification_report(y_test, y_pred))
```
